```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("/content/Housing.csv")
```

```python
X = np.array(data['area'])
Y = np.array(data['price'])
N = len(X)

sum_X = np.sum(X)
sum_Y = np.sum(Y)
sum_XY = np.sum(X*Y)
sum_X2 = np.sum(X**2)

m = (N * sum_XY - sum_X * sum_Y)/(N * sum_X2 - sum_X * sum_X**2)
c = (sum_Y - m * sum_X)/N

Y_pred = m * X + c

plt.scatter(X, Y, color='red', label="Data Points")
plt.plot(X, Y_pred, color='blue', label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")
```
