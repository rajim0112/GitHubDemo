# Task 1: Covering Basic Pandas

```python
import pandas as pd
```

```python
data = pd.read_csv("/content/generated_dataset.csv")
```

```python
data.isnull().sum()
```

```python
# Filter rows where Salary > 50000
filtered_data = data[data['Salary'] > 50000]
print("\nRows where Salary > 50000:")
print(filtered_data)
```

```python
# Select specific columns
selected_columns = data[['Name', 'Salary']]
print("\nSelected columns (Name, Salary):")
print(selected_columns)
```

```python
# Group by Department and calculate the average salary
average_salary_by_department = data.groupby('Department')['Salary'].mean()
print("\nAverage salary by department:")
print(average_salary_by_department)
```

```python
# Count the number of employees in each department
employee_count_by_department = data['Department'].value_counts()
print("\nNumber of employees in each department:")
print(employee_count_by_department)
```

```python
# Handling missing values
data_cleaned = data.fillna({'Salary': 0})
print("\nDataset after handling missing values:")
print(data_cleaned)
```

```python
# Renaming salary columns
data_renamed = data.rename(columns={'Name': 'Employee Name', 'Salary': 'Monthly Salary'})
print("\nDataset after renaming columns:")
print(data_renamed)
```

```python
# Adding Annual Salary column
data['Annual Salary'] = data['Salary'] * 12
print("\nDataset with new column 'Annual Salary':")
print(data)
```
