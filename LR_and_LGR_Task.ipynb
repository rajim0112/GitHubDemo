Task 1: Linear Regression (Predict House Prices)


```python
import pandas as pd

# Sample Data (Dataset for House Prices)
data = {
    'Square_Feet': [1500, 1800, 1200, 2000],
    'Bedrooms': [3, 4, 2, 4],
    'Bathrooms': [2, 3, 1, 3],
    'Location_Score': [8.5, 9.0, 7.2, 9.5],
    'Price': [250000, 320000, 180000, 350000]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Check for missing values
print(df.isnull().sum())

# Normalize features (if necessary)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Square_Feet', 'Bedrooms', 'Bathrooms', 'Location_Score']])

# Combine the scaled features with the target variable (Price)
X = pd.DataFrame(scaled_features, columns=['Square_Feet', 'Bedrooms', 'Bathrooms', 'Location_Score'])
y = df['Price']
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the training data
y_pred = model.predict(X)

# Calculate R² score
r2 = r2_score(y, y_pred)
print(f'R² score: {r2}')

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
# New house data
new_house = [[1800, 3, 2, 8.7]]  # Example new house data

# Normalize the new data (using the same scaler)
scaled_new_house = scaler.transform(new_house)

# Predict the price
predicted_price = model.predict(scaled_new_house)
print(f'Predicted Price: ${predicted_price[0]:,.2f}')
```

Task 2: Logistic Regression (Customer Churn
Prediction)

```python
import pandas as pd

# Sample Data (Customer Churn)
data = {
    'Monthly_Charges': [50, 80, 40, 90],
    'Contract_Type': [0, 1, 0, 1],
    'Tenure': [12, 24, 6, 36],
    'Support_Requests': [2, 0, 3, 1],
    'Churn': [1, 0, 1, 0]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Separate features and target
X = df[['Monthly_Charges', 'Contract_Type', 'Tenure', 'Support_Requests']]
y = df['Churn']

# Normalize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Show the scaled features
print(X_scaled)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
model = LogisticRegression()

# Train the model
model.fit(X_scaled, y)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

# Make predictions on the data
y_pred = model.predict(X_scaled)

# Evaluate using different metrics
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
conf_matrix = confusion_matrix(y, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'Confusion Matrix:\n{conf_matrix}')
```

```python
# New customer data: [Monthly_Charges, Contract_Type, Tenure, Support_Requests]
new_customer = [[70, 0, 18, 1]]  # Example new customer data

# Normalize the new data
scaled_new_customer = scaler.transform(new_customer)

# Predict the churn (0=No, 1=Yes)
churn_prediction = model.predict(scaled_new_customer)
print(f'Will the customer churn? {"Yes" if churn_prediction[0] == 1 else "No"}')
```
